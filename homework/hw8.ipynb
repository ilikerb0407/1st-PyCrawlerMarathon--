```python
import requests
url='https://www.zhihu.com/explore'
headers = {'user-agent':'abc'}
r = requests.get(url ,headers=headers)
r.encoding = 'utf-8'
print (r.text[0:600])
```

```python
response = r.text
soup = BeautifulSoup(response, 'html5lib')
print (soup)
```
